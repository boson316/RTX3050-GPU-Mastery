# 01 — Getting Started on RTX 3050

This notebook checks your GPU and runs a few PyTorch CUDA operations. **Run on an RTX 3050** (or any CUDA GPU) for best alignment with this lab.

```python
import torch

print("PyTorch:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("Device:", torch.cuda.get_device_name(0))
    print("Memory (GB):", round(torch.cuda.get_device_properties(0).total_memory / 1e9, 2))
```

## First GPU ops: vector add and matmul

```python
if torch.cuda.is_available():
    n = 1_000_000
    a = torch.randn(n, device="cuda", dtype=torch.float32)
    b = torch.randn(n, device="cuda", dtype=torch.float32)
    torch.cuda.synchronize()
    t0 = torch.cuda.Event(enable_timing=True)
    t1 = torch.cuda.Event(enable_timing=True)
    t0.record()
    c = a + b
    t1.record()
    torch.cuda.synchronize()
    print(f"Vector add {n} elements: {t0.elapsed_time(t1):.4f} ms")
    del a, b, c
    torch.cuda.empty_cache()
```

```python
if torch.cuda.is_available():
    N = 1024
    A = torch.randn(N, N, device="cuda", dtype=torch.float16)
    B = torch.randn(N, N, device="cuda", dtype=torch.float16)
    for _ in range(5):
        _ = torch.matmul(A, B)
    torch.cuda.synchronize()
    t0 = torch.cuda.Event(enable_timing=True)
    t1 = torch.cuda.Event(enable_timing=True)
    t0.record()
    for _ in range(50):
        _ = torch.matmul(A, B)
    t1.record()
    torch.cuda.synchronize()
    ms = t0.elapsed_time(t1) / 50
    gflops = 2 * N**3 / 1e9 / (ms / 1000)
    print(f"Matmul {N}x{N} FP16: {ms:.4f} ms, {gflops:.1f} GFLOPS")
else:
    print("No CUDA; run on GPU for timing.")
```
